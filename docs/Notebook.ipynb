```python
import pandas as pd
import itertools

data = pd.read_json("data/artists.json")
```

```python
track_values = data.tracks.values.flatten()
tracks = list(itertools.chain(*track_values))
tdf = pd.DataFrame(data=tracks)
```

```python
tdf.date = pd.to_datetime(tdf.date)
tdf.index = tdf.date
tdf.plot.line(x="date", title="Plays Over Time")
```
